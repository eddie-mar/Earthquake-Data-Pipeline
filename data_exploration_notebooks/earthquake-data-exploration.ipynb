```python
df_mag_null.sample(10)
```

```python
df_mag_null.count()
```

```python
df_mag_null.type.value_counts()
```

```python
df_mag_null.describe()
```

Although data for null magnitude has a lot of valid depths, it is difficult to assume a value and the number of data is significantly large to let it to 0, thus, we shall remove this null values.

```python
df_dep_null = df[df.depth.isnull()]
```

```python
df.loc[df['depth'] == 0].describe()
```

Data with null depth most has valid magnitude. I looked at the whole data with 0 depth and it has even lower mean magnitude than the null depth data. There is a significant number of 0 depth, thus, it is a valid value. I shall assume 0 for the null depth as it can be a human error to write none or nothing for a 0 value.
